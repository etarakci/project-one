```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from functools import reduce
import datetime as dt
```

```python
#read CSV files
df1=pd.read_csv("data-APZBu.csv") #Nasdaq_data
df2=pd.read_csv("data-EKHTG.csv")#Dow_data
df3=pd.read_csv("data-ULLTa.csv") #S&P500_data

# compile the list of dataframes you want to merge
data_frames = [df1, df2, df3]

#merge our data frames
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'],
                                            how='outer'), data_frames)
#merge our data frames
#combined_data=pd.merge(Nasdaq_data,Dow_data,SP500_data, on="left")

df_merged
```

```python
df_merged.info()
```

```python
df_merged
```

```python
# Plot all the close prices
df_merged['Date']=pd.to_datetime (df_merged ['Date'])
df_merged.set_index('Date', inplace=True)
df_merged.plot( figsize=(10, 7))
# Show the legend
plt.legend()
# Define the label for the title of the figure
plt.title(" Close Price Index", fontsize=16)
# Define the labels for x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Date', fontsize=14)
# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
